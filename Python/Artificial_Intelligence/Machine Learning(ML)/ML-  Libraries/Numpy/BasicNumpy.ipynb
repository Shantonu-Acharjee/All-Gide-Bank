## **Multidimensional Arrays in Python**

```python
from numpy import *
```

```python
array1 = array([
                [1,2,3,4,6,8],
                [8,4,3,6,2,1]
])

print(array1)
```

```python
print(array1.dtype) # show data type
```

```python
print(array1.ndim) # show dymanshon
```

```python
print(array1.shape) # show array shape
```

```python
print(array1.size)
```

```python
array2 = array1.flatten() # 2 dymanshon to 1 dymanshon
print(array2)
```

```python
array3 = array2.reshape(2,6) # 1 dymonshon to 2 dymanshon
print(array3)
```

```python
array4 = array2.reshape(2,2,3) # make 3 dymonshon array
print(array4) 
```

```python
print(array4.ndim)
```

# **Matrix Multiplication**

```python
m1 = ([1,2,3],
      [4,5,6],
      [7,3,5]
      )

m2 = ([3,4,2],
      [7,8,9],
      [3,5,9]
      )

print(m1)
print(m2)
```

```python
m1 = ([1,2,3],
      [4,5,6],
      [7,3,5]
      )

m2 = ([3,4,2],
      [7,8,9],
      [3,5,9]
      )

result = np.add(m1,m2) # addition
print(result)
```

```python
m1 = ([1,2,3],
      [4,5,6],
      [7,3,5]
      )

m2 = ([3,4,2],
      [7,8,9],
      [3,5,9]
      )

result = np.subtract(m1,m2) # subtractshon
print(result)
```

```python
m1 = ([1,2,3],
      [4,5,6],
      [7,3,5]
      )

m2 = ([3,4,2],
      [7,8,9],
      [3,5,9]
      )

result = np.multiply(m1,m2) # subtractshon
print(result)
```

```python
m1 = ([1,2,3],
      [4,5,6],
      [7,3,5]
      )

m2 = ([3,4,2],
      [7,8,9],
      [3,5,9]
      )
# 1*3 + 2*7 + 3*3 = 26
result = np.dot(m1,m2)
print(result)
```
